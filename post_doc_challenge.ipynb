```python
import torch
import numpy as np
import medmnist
from medmnist import INFO
from torchvision import transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from tqdm import tqdm
import torchvision.models as models
import torch.nn as nn
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

This block of code is setting up the PneumoniaMNIST dataset for training a model like ResNet

```python
# Dataset selection
data_flag = 'pneumoniamnist'
info = INFO[data_flag]
DataClass = getattr(medmnist, info['python_class'])

# Transform images to 3 channels and resize
transform = transforms.Compose([
    transforms.Resize((224, 224)),              # Resize for ResNet
    transforms.Grayscale(num_output_channels=3), # Convert 1 channel -> 3 channels
    transforms.ToTensor()                       # Convert to tensor
])

# Load datasets
train_dataset = DataClass(split='train', transform=transform, download=True)
val_dataset = DataClass(split='val', transform=transform, download=True)
test_dataset = DataClass(split='test', transform=transform, download=True)

print("Train images:", len(train_dataset))
print("Validation images:", len(val_dataset))
print("Test images:", len(test_dataset))
```

This code is preparing ResNet18 as a feature extractor (not for classification yet)

```python
# Load pretrained ResNet18
model = models.resnet18(pretrained=True)

# Remove the final fully connected layer
model = nn.Sequential(*list(model.children())[:-1])

# Move to device and set to evaluation mode
model = model.to(device)
model.eval()

print("ResNet18 loaded for embedding extraction!")
```

**This function is for extracting feature embeddings from a dataset using the pretrained ResNet18.**

```python
def extract_embeddings(dataset): #Function definition
    loader = DataLoader(dataset, batch_size=32, shuffle=False) #Create a DataLoader
    embeddings = [] #Initialize lists
    labels = []

    with torch.no_grad(): #Disable gradient computation
        for images, targets in tqdm(loader):#Loop over DataLoader
            images = images.to(device) #Move images to the device
            output = model(images)#Pass images through ResNet
            output = output.view(output.size(0), -1)  # flatten 512-d vector
            embeddings.append(output.cpu().numpy())#Store embeddings and labels
            labels.append(targets.numpy())

    return np.vstack(embeddings), np.vstack(labels)#Combine all batches
```

**This code is using the extract_embeddings function you defined to get feature vectors and labels for the test dataset.**

```python
test_embeddings, test_labels = extract_embeddings(test_dataset)#Extract embeddings

print("Embeddings shape:", test_embeddings.shape)#Print shapes
print("Labels shape:", test_labels.shape)
```

**This code is building a FAISS index for your test embeddings, which allows fast nearest neighbor search**

```python
import faiss #Import FAISS --FAISS is a library from Facebook AI

# Dimension of embeddings
dimension = test_embeddings.shape[1]

# Create index (L2 distance)
index = faiss.IndexFlatL2(dimension)

# Add embeddings to the index
index.add(test_embeddings)

print("FAISS index built. Total vectors:", index.ntotal)
```

**This function is for retrieving the top-k most similar images from the FAISS index using a single query embedding**

```python
def search(query_embedding, k=5):#Function definition
    """
    Input:
        query_embedding: numpy array of shape (512,)
        k: number of top similar images to retrieve
    Output:
        indices of top-k similar images in the test set
    """
    distances, indices = index.search(query_embedding.reshape(1, -1), k)#Reshape the query
    return indices[0]
```

**This function is for visualizing a query image and its top-k retrieved images from the FAISS search**

```python
def show_results(query_id, retrieved_ids):#Function definition
    """
    Displays the query image and top-k retrieved images.
    """
    plt.figure(figsize=(12,3))

    # Show query image
    img, _ = test_dataset[query_id]
    plt.subplot(1, len(retrieved_ids)+1, 1)
    plt.imshow(img.permute(1,2,0))  # convert tensor to HWC for RGB display
    plt.title("Query")
    plt.axis('off')

    # Show retrieved images
    for i, idx in enumerate(retrieved_ids):
        img, _ = test_dataset[idx]
        plt.subplot(1, len(retrieved_ids)+1, i+2)
        plt.imshow(img.permute(1,2,0))
        plt.title(f"Top {i+1}")
        plt.axis('off')

    plt.show()
```

```python
# Choose a query image index
query_id = 0
query_embedding = test_embeddings[query_id]

# Retrieve top-5 similar images
top_k = search(query_embedding, k=5)
print("Top-5 retrieved indices:", top_k)

# Visualize
show_results(query_id, top_k)
```

**This function calculates Precision@k for your test set using the embeddings and FAISS search**

```python
def precision_at_k(k=5):#Function definition
    correct = 0 # Initialize counters
    total = len(test_embeddings)

    for i in range(total): #Loop over all test samples
        query_label = test_labels[i]
        retrieved = search(test_embeddings[i], k=k)

        match = 0 #Count correct matches
        for idx in retrieved:
            if test_labels[idx] == query_label:
                match += 1

        correct += match / k #Compute precision for this query

    return correct / total #Return average precision

print("Precision@5:", precision_at_k(5))
```

```python
import torch
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from medmnist import PneumoniaMNIST
```

```python
train_dataset = PneumoniaMNIST(split='train', transform=train_transform, download=True)
val_dataset = PneumoniaMNIST(split='val', transform=test_transform, download=True)
test_dataset = PneumoniaMNIST(split='test', transform=test_transform, download=True)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d(1)
        )

        self.classifier = nn.Linear(128, 1)

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x
```

```python
model = SimpleCNN()
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)
```

```python
def evaluate(model, loader):
    model.eval()
    preds, probs, labels_list = [], [], []

    with torch.no_grad():
        for images, labels in loader:
            outputs = model(images).squeeze()
            probability = torch.sigmoid(outputs)

            preds.extend((probability > 0.5).int().numpy())
            probs.extend(probability.numpy())
            labels_list.extend(labels.numpy())

    return np.array(preds), np.array(probs), np.array(labels_list)
```

```python
preds, probs, labels = evaluate(model, test_loader)

print(classification_report(labels, preds))
```

```python
cm = confusion_matrix(labels, preds)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
auc = roc_auc_score(labels, probs)
fpr, tpr, _ = roc_curve(labels, probs)

plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
import torch
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from medmnist import PneumoniaMNIST
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# 1️⃣ Setup device
# -------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# -------------------------------
# 2️⃣ Transforms & Dataset
# -------------------------------
train_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[.5], std=[.5])
])
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[.5], std=[.5])
])

train_dataset = PneumoniaMNIST(split='train', transform=train_transform, download=True)
val_dataset   = PneumoniaMNIST(split='val', transform=test_transform, download=True)
test_dataset  = PneumoniaMNIST(split='test', transform=test_transform, download=True)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=64, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=64, shuffle=False)

# -------------------------------
# 3️⃣ Load your trained model
# -------------------------------
# Example: a pretrained model placeholder
# Replace this with your actual model
import torch.nn as nn
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 16, 3, 1, 1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(16*14*14, 2)  # PneumoniaMNIST images are 28x28 grayscale
        )
    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        return x

model = SimpleCNN().to(device)
# model.load_state_dict(torch.load('your_model.pth'))  # Load trained weights if available
model.eval()

# -------------------------------
# 4️⃣ Evaluation: y_true, y_pred
# -------------------------------
y_true = []
y_pred = []
y_scores = []  # for ROC-AUC

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(outputs, dim=1)

        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())
        y_scores.extend(probs[:,1].cpu().numpy())  # probability for class 1

# -------------------------------
# 5️⃣ Classification Report
# -------------------------------
report_dict = classification_report(y_true, y_pred, output_dict=True)
df_report = pd.DataFrame(report_dict).transpose()
print("✅ Classification Report:\n")
print(df_report.to_markdown())

# -------------------------------
# 6️⃣ Confusion Matrix
# -------------------------------
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(5,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal','Pneumonia'], yticklabels=['Normal','Pneumonia'])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

# -------------------------------
# 7️⃣ ROC Curve & AUC
# -------------------------------
roc_auc = roc_auc_score(y_true, y_scores)
fpr, tpr, thresholds = roc_curve(y_true, y_scores)

plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
